```python
import pandas as pd

# 파일 불러오기
au_sept = pd.read_csv('./9월_AU.csv')
revenue_sept = pd.read_csv('./9월_매출.csv')
au_aug = pd.read_csv('./8월_AU.csv')
revenue_aug = pd.read_csv('./8월_매출.csv')

# 9월 데이터 병합 (동시접속자 & 매출)
sept_data = pd.merge(au_sept, revenue_sept, on='게임명')

# 8월 데이터 병합 (동시접속자 & 매출)
aug_data = pd.merge(au_aug, revenue_aug, on='게임명')

# 9월과 8월 데이터에 대해 상관관계 분석 (달러 매출)
sept_corr = sept_data[['9월 MAU(만)', '9월 달러 매출 추정(만)']].corr().iloc[0, 1]
aug_corr = aug_data[['8월 MAU(만)', '8월 달러 매출 추정(만)']].corr().iloc[0, 1]

# 9월과 8월 데이터에 대해 상관관계 분석 (원 매출)
sept_corr_krw = sept_data[['9월 MAU(만)', '9월 원 매출 추정(억)']].corr().iloc[0, 1]
aug_corr_krw = aug_data[['8월 MAU(만)', '8월 원 매출 추정(억)']].corr().iloc[0, 1]

# 결과 출력
print(f"9월 상관계수 (달러 매출): {sept_corr:.2f}")
print(f"8월 상관계수 (달러 매출): {aug_corr:.2f}")
print(f"9월 상관계수 (원화 매출): {sept_corr_krw:.2f}")
print(f"8월 상관계수 (원화 매출): {aug_corr_krw:.2f}")
```

```python
# 데이터 병합 후 증감율 계산
merged_data = pd.merge(aug_data, sept_data, on='게임명', how='outer', suffixes=('_8월', '_9월'))

# 증감율 계산
merged_data['MAU 증감율 (%)'] = ((merged_data['9월 MAU(만)'] - merged_data['8월 MAU(만)']) / merged_data['8월 MAU(만)']) * 100
merged_data['달러 매출 증감율 (%)'] = ((merged_data['9월 달러 매출 추정(만)'] - merged_data['8월 달러 매출 추정(만)']) / merged_data['8월 달러 매출 추정(만)']) * 100
merged_data['원화 매출 증감율 (%)'] = ((merged_data['9월 원 매출 추정(억)'] - merged_data['8월 원 매출 추정(억)']) / merged_data['8월 원 매출 추정(억)']) * 100

merged_data[['게임명', '9월 MAU(만)', '8월 MAU(만)', 'MAU 증감율 (%)', 
             '9월 달러 매출 추정(만)', '8월 달러 매출 추정(만)', '달러 매출 증감율 (%)',
             '9월 원 매출 추정(억)', '8월 원 매출 추정(억)', '원화 매출 증감율 (%)']]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
```

```python
top_changes = merged_data.sort_values(by='MAU 증감율 (%)', ascending=False)

# Figure 설정
fig, axes = plt.subplots(3, 1, figsize=(20, 18))

# 각 증감율 그래프 그리기
# 1. MAU 증감율
axes[0].bar(top_changes['게임명'], top_changes['MAU 증감율 (%)'], color='skyblue')
axes[0].set_title('8월 대비 9월 MAU 증감율 (%)')
axes[0].set_xlabel('게임명')
axes[0].set_ylabel('MAU 증감율 (%)')
axes[0].tick_params(axis='x', rotation=45)

# 2. 달러 매출 증감율
axes[1].bar(top_changes['게임명'], top_changes['달러 매출 증감율 (%)'], color='salmon')
axes[1].set_title('8월 대비 9월 달러 매출 증감율 (%)')
axes[1].set_xlabel('게임명')
axes[1].set_ylabel('달러 매출 증감율 (%)')
axes[1].tick_params(axis='x', rotation=45)

# 3. 원화 매출 증감율
axes[2].bar(top_changes['게임명'], top_changes['원화 매출 증감율 (%)'], color='lightgreen')
axes[2].set_title('8월 대비 9월 원화 매출 증감율 (%)')
axes[2].set_xlabel('게임명')
axes[2].set_ylabel('원화 매출 증감율 (%)')
axes[2].tick_params(axis='x', rotation=45)

# 레이아웃 조정
plt.tight_layout()
plt.show()
```
