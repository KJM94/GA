```python
import matplotlib.pyplot as plt
import pandas as pd

import seaborn as sns

plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False

#  
data = {
    "Category": ["MAU", " ", " "],
    "Values": [59, 119, 416],  # : MAU(), ( )
}

regional_sales = {
    "Region": ["", "", "", "( )"],
    "Percentage": [31.85, 28.64, 19.82, 19.69],
}

user_data = {
    "Category": ["", "20 "],
    "Percentage": [92.71, 50],
}

# 1. MAU    ( )
plt.figure(figsize=(8, 6))
plt.bar(data["Category"], data["Values"], color=["blue", "green", "orange"])
plt.title("2024.12 MAU   ")
plt.ylabel("(MAU) /  ()")
plt.xticks(rotation=0)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()

# 2.    ( )
plt.figure(figsize=(8, 6))
plt.pie(regional_sales["Percentage"], labels=regional_sales["Region"], autopct="%.2f%%", startangle=90, colors=["gold", "lightblue", "lightcoral", "lightgreen"])
plt.title("2024.12   ")
plt.gca().add_artist(plt.Circle((0, 0), 0.7, color="white"))  #   
plt.tight_layout()
plt.show()

# 3.      (  )
plt.figure(figsize=(8, 6))
plt.bar(user_data["Category"], user_data["Percentage"], color=["purple", "pink"])
plt.title("2024.12     ")
plt.ylabel(" (%)")
plt.ylim(0, 100)
plt.xticks(rotation=0)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
#  
domestic_mau = 59  #  MAU (:  )
domestic_revenue = 119  #   (:  )
global_revenue = 416  #   (:  )
regional_revenue_percentage = {"": 31.85, "": 28.64, "": 19.82, "": 19.69}

#  ARPU 
domestic_arpu = domestic_revenue * 1e8 / (domestic_mau * 1e4)  # : /

#  MAU  (ARPU   )
regional_mau_estimates = {}
for region, percentage in regional_revenue_percentage.items():
    revenue = percentage / 100 * global_revenue  #   ( )
    mau = revenue * 1e8 / domestic_arpu / 1e4  # MAU  ( )
    regional_mau_estimates[region] = mau

# :  ARPU   MAU
regions = list(regional_mau_estimates.keys())
mau_values = list(regional_mau_estimates.values())

plt.figure(figsize=(8, 6))
plt.bar([""] + regions, [domestic_mau] + mau_values, color=["blue", "green", "orange", "red", "purple"])
plt.title("2024.12  MAU  (:  )")
plt.ylabel("MAU ( )")
plt.xlabel("")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()

#  ARPU 
print(f" ARPU: {domestic_arpu:,.0f} /")
```
