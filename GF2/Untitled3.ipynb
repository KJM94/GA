```python
import matplotlib.pyplot as plt
import pandas as pd

import seaborn as sns

plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False

# 데이터 입력
data = {
    "Category": ["MAU", "국내 매출", "글로벌 매출"],
    "Values": [59, 119, 416],  # 단위: MAU(백만), 매출(억 원)
}

regional_sales = {
    "Region": ["일본", "한국", "북미", "기타(중국 포함)"],
    "Percentage": [31.85, 28.64, 19.82, 19.69],
}

user_data = {
    "Category": ["남성", "20대 남성"],
    "Percentage": [92.71, 50],
}

# 1. MAU 및 매출 데이터 (막대 그래프)
plt.figure(figsize=(8, 6))
plt.bar(data["Category"], data["Values"], color=["blue", "green", "orange"])
plt.title("2024.12 MAU 및 매출 데이터")
plt.ylabel("백만(MAU) / 억 원(매출)")
plt.xticks(rotation=0)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()

# 2. 글로벌 매출 비중 (도넛 차트)
plt.figure(figsize=(8, 6))
plt.pie(regional_sales["Percentage"], labels=regional_sales["Region"], autopct="%.2f%%", startangle=90, colors=["gold", "lightblue", "lightcoral", "lightgreen"])
plt.title("2024.12 글로벌 매출 비중")
plt.gca().add_artist(plt.Circle((0, 0), 0.7, color="white"))  # 도넛 차트 효과
plt.tight_layout()
plt.show()

# 3. 사용자 성별 및 연령 데이터 (스택형 막대 그래프)
plt.figure(figsize=(8, 6))
plt.bar(user_data["Category"], user_data["Percentage"], color=["purple", "pink"])
plt.title("2024.12 사용자 성별 및 연령 데이터")
plt.ylabel("비율 (%)")
plt.ylim(0, 100)
plt.xticks(rotation=0)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# 데이터 입력
domestic_mau = 59  # 국내 MAU (단위: 만 명)
domestic_revenue = 119  # 국내 매출 (단위: 억 원)
global_revenue = 416  # 글로벌 매출 (단위: 억 원)
regional_revenue_percentage = {"일본": 31.85, "한국": 28.64, "북미": 19.82, "기타": 19.69}

# 국내 ARPU 계산
domestic_arpu = domestic_revenue * 1e8 / (domestic_mau * 1e4)  # 단위: 원/명

# 지역별 MAU 추정 (ARPU가 국내와 유사하다고 가정)
regional_mau_estimates = {}
for region, percentage in regional_revenue_percentage.items():
    revenue = percentage / 100 * global_revenue  # 지역별 매출 (억 원)
    mau = revenue * 1e8 / domestic_arpu / 1e4  # MAU 추정 (만 명)
    regional_mau_estimates[region] = mau

# 시각화: 국내 ARPU와 지역별 추정 MAU
regions = list(regional_mau_estimates.keys())
mau_values = list(regional_mau_estimates.values())

plt.figure(figsize=(8, 6))
plt.bar(["한국"] + regions, [domestic_mau] + mau_values, color=["blue", "green", "orange", "red", "purple"])
plt.title("2024.12 지역별 MAU 추정 (단위: 만 명)")
plt.ylabel("MAU (만 명)")
plt.xlabel("지역")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()

# 국내 ARPU 출력
print(f"국내 ARPU: {domestic_arpu:,.0f} 원/명")
```

```python
# 데이터 입력
regions = ["일본", "한국", "북미", "기타"]
revenue_percentage = [31.85, 28.64, 19.82, 19.69]
revenue_values = [416 * p / 100 for p in revenue_percentage]  # 지역별 매출 계산 (억 원)

# 파이 차트: 지역별 매출 비중
plt.figure(figsize=(6, 6))
plt.pie(revenue_percentage, labels=regions, autopct='%1.1f%%', startangle=90, colors=["gold", "blue", "green", "red"])
plt.title("2024.12 지역별 매출 비중")
plt.show()

# 막대 그래프: 지역별 매출
plt.figure(figsize=(8, 6))
plt.bar(regions, revenue_values, color=["gold", "blue", "green", "red"])
plt.title("2024.12 지역별 매출 (억 원)")
plt.ylabel("매출 (억 원)")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```
